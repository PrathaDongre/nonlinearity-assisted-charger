```python
index_where_1 = np.array([1224, 1225, 1274, 1275])

CHI00_beta_zero_v1 = chiR00_T_v1[:,index_where_1[0]]+chiI00_T_v1[:,index_where_1[0]]*1.0j
CHI01_beta_zero_v1 = chiR01_T_v1[:,index_where_1[0]]+chiI01_T_v1[:,index_where_1[0]]*1.0j
CHI10_beta_zero_v1 = chiR10_T_v1[:,index_where_1[0]]+chiI10_T_v1[:,index_where_1[0]]*1.0j
CHI11_beta_zero_v1 = chiR11_T_v1[:,index_where_1[0]]+chiI11_T_v1[:,index_where_1[0]]*1.0j

CHI00_beta_zero_v2 = chiR00_T_v2[:,index_where_1[0]]+chiI00_T_v2[:,index_where_1[0]]*1.0j
CHI01_beta_zero_v2 = chiR01_T_v2[:,index_where_1[0]]+chiI01_T_v2[:,index_where_1[0]]*1.0j
CHI10_beta_zero_v2 = chiR10_T_v2[:,index_where_1[0]]+chiI10_T_v2[:,index_where_1[0]]*1.0j
CHI11_beta_zero_v2 = chiR11_T_v2[:,index_where_1[0]]+chiI11_T_v2[:,index_where_1[0]]*1.0j

CHI00_beta_zero_v3 = chiR00_T_v3[:,index_where_1[0]]+chiI00_T_v3[:,index_where_1[0]]*1.0j
CHI01_beta_zero_v3 = chiR01_T_v3[:,index_where_1[0]]+chiI01_T_v3[:,index_where_1[0]]*1.0j
CHI10_beta_zero_v3 = chiR10_T_v3[:,index_where_1[0]]+chiI10_T_v3[:,index_where_1[0]]*1.0j
CHI11_beta_zero_v3 = chiR11_T_v3[:,index_where_1[0]]+chiI11_T_v3[:,index_where_1[0]]*1.0j


CHI00_beta_zero_lin_v8 = chiR00_T_lin_v8[:,index_where_1[0]]+chiI00_T_lin_v8[:,index_where_1[0]]*1.0j
CHI01_beta_zero_lin_v8 = chiR01_T_lin_v8[:,index_where_1[0]]+chiI01_T_lin_v8[:,index_where_1[0]]*1.0j
CHI10_beta_zero_lin_v8 = chiR10_T_lin_v8[:,index_where_1[0]]+chiI10_T_lin_v8[:,index_where_1[0]]*1.0j
CHI11_beta_zero_lin_v8 = chiR11_T_lin_v8[:,index_where_1[0]]+chiI11_T_lin_v8[:,index_where_1[0]]*1.0j
```

```python
def abs2(x):
    return x.real**2 + x.imag**2

omega_0 = 1
E_B_v1 = omega_0*CHI11_beta_zero_v1
ergotr_B_by_omega_v1 = (0.5)*(np.sqrt(np.square(CHI11_beta_zero_v1-CHI00_beta_zero_v1)+4*abs2(CHI10_beta_zero_v1))+CHI11_beta_zero_v1-CHI00_beta_zero_v1)

E_B_v2 = omega_0*CHI11_beta_zero_v2
ergotr_B_by_omega_v2 = (0.5)*(np.sqrt(np.square(CHI11_beta_zero_v2-CHI00_beta_zero_v2)+4*abs2(CHI10_beta_zero_v2))+CHI11_beta_zero_v2-CHI00_beta_zero_v2)

E_B_v3 = omega_0*CHI11_beta_zero_v3
ergotr_B_by_omega_v3 = (0.5)*(np.sqrt(np.square(CHI11_beta_zero_v3-CHI00_beta_zero_v3)+4*abs2(CHI10_beta_zero_v3))+CHI11_beta_zero_v3-CHI00_beta_zero_v3)

E_B_lin_v8 = omega_0*CHI11_beta_zero_lin_v8
ergotr_B_by_omega_lin_v8 = (0.5)*(np.sqrt(np.square(CHI11_beta_zero_lin_v8-CHI00_beta_zero_lin_v8)+4*abs2(CHI10_beta_zero_lin_v8))+CHI11_beta_zero_lin_v8-CHI00_beta_zero_lin_v8)
```

```python
plt.plot(ergotr_B_by_omega_v1[:390])
plt.plot(ergotr_B_by_omega_v2[:390])
plt.plot(ergotr_B_by_omega_v3[:390])
plt.title("comparison of the different versions of non-lin runs")
plt.show()
```
